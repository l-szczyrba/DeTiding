## Try Fast Fourier Transform - SP

Calculate fast fourier transform of the total water level signal

```python
fft_vals = fft(df_SP['N_Obs_m'])
```

Get the power spectral density by taking the square of the absolute value of fft_vals

```python
psd = np.abs(fft_vals) ** 2
```

Calculate the frequencies

Method #2 (https://ipython-books.github.io/101-analyzing-the-frequency-components-of-a-signal-with-a-fast-fourier-transform/_) \

Day as frequency unit, aka frequency of 1 corresponsd to 1 day

```python
freqs = fftfreq(len(psd), 1. / 24 )
mask = freqs > 0
```

## Plots

```python
y = df_SP['N_Obs_m']
```

```python
sns.set(rc={'figure.figsize':(11, 4)})
```

```python
plt.figure(1)
plt.title ('NOAA Measured Total Water Level')
plt.plot(x, y, label = 'Original TWL')
plt.legend()
```


Power spectral density of signal as a function of the frequency (in unit of 1/week) \
Lograithmic scale for y axis (decibels)

```python
fig, ax = plt.subplots(1, 1, figsize=(11, 4))
ax.plot(freqs[mask], 10 * np.log10(psd[mask]))
ax.set_xlim(0, 5)
ax.set_xlabel('Frequency (1/day)')
ax.set_ylabel('PSD (dB)')
```

Cut out frequencies higher than the fundemental frequency using the daily frequency measure

```python
fft_bis = fft_vals.copy()
fft_bis[np.abs(freqs) > 1.8] = 0
```

```python
detide = np.real(ifft(fft_bis))
```

```python
plt.figure(1)
plt.title ('NOAA Measured Total Water Level')
plt.plot(x, y, label = 'Original TWL')
plt.legend()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(11, 4))
ax.plot(x, detide, linewidth = 1)
ax.set_title('Non-Tidal Component by FFT')
ax.set_xlabel('Date')
ax.set_ylabel('Water Level (m)')
```
